### 1Ô∏è‚É£ Importing Required Libraries

- Analysis: Import and Load dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set visualization style
sns.set(style="whitegrid", palette="Set2")
```

### 2Ô∏è‚É£ Loading the Dataset

```python
# Load dataset
data = pd.read_csv(r"C:\Users\Dell\OneDrive\Saiket Systems\Telco_Customer_Churn_Dataset.csv")
data
```

```python
# Display first few rows
print("üîπ First 5 Rows of the Dataset:")
print(data.head(5))
```

```python
# Check dataset size and missing values
print("\nüîπ Dataset Shape:", data.shape)
print("\nüîπ Missing Values:\n", data.isnull().sum())
```

### 3Ô∏è‚É£ Churn Rate Calculation

- Analysis: Churn rate calculation
- Purpose: Measure overall churn percentage
- Visualization: Pie chart

-  "Churn" column indicates whether the customer left or stayed (Yes/No)

```python
# Calculate churn rate as percentage
churn_rate = data['Churn'].value_counts(normalize=True) * 100
print("\nüîπ Churn Rate (%):\n", churn_rate)
```

```python
# ‚úÖ Visualization: Overall Churn Rate
plt.figure(figsize=(5, 5))
plt.pie(data['Churn'].value_counts(),
        labels=['No (Stayed)', 'Yes (Churned)'],
        autopct='%1.1f%%',
        colors=['#66b3ff', '#ff9999'])
plt.title('Overall Customer Churn Rate')
plt.show()
```

### 4Ô∏è‚É£ Customer Demographics Analysis

- Analysis: Demographic analysis
- Purpose: Understand churn by gender, partner, dependents
- Visualization: Bar charts

```python
# Gender vs Churn
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='gender', hue='Churn')
plt.title('Churn Distribution by Gender')
plt.show()
```

```python
# Partner Status vs Churn
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Partner', hue='Churn')
plt.title('Churn Distribution by Partner Status')
plt.show()
```

```python
# Dependent Status vs Churn
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Dependents', hue='Churn')
plt.title('Churn Distribution by Dependent Status')
plt.show()
```

### 5Ô∏è‚É£ Tenure Analysis

- Analysis: Tenure analysis
- Purpose: See how customer duration affects churn
- Visualization: Histogram, Box plot

- Tenure = number of months a customer has stayed.

```python
# Tenure Distribution
plt.figure(figsize=(7, 4))
sns.histplot(data['tenure'], bins=30, kde=True)
plt.title('Distribution of Customer Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Tenure vs Churn
plt.figure(figsize=(7, 4))
sns.boxplot(data=data, x='Churn', y='tenure')
plt.title('Tenure vs Churn')
plt.show()
```

### 6Ô∏è‚É£ Contract Type and Payment Method Analysis

- Analysis: Contract & payment
- Purpose: Study how plan and payment type affect churn
- Visualization: Bar charts

```python
# Churn by Contract Type
plt.figure(figsize=(7, 4))
sns.countplot(data=data, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.show()
```

```python
# Churn by Payment Method
plt.figure(figsize=(8, 4))
sns.countplot(data=data, x='PaymentMethod', hue='Churn')
plt.title('Churn by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

### 7Ô∏è‚É£ Summary Statistics and Correlation 

- Analysis: Correlation
- Purpose: Identify numeric relationships
- Visualization: Heat map

```python
print("\nüîπ Summary Statistics for Tenure:")
print(data['tenure'].describe())
```

```python
# Correlation Heatmap (for numeric columns)
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
